```python
# loading relevant packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Exploratory Data Analysis  
- Load in the cleaned data

```python
df = pd.read_csv('cleaned_aviation_data.csv')
df.head(10)
```

```python
df.info()
```

## Explore safety metrics across models/makes
- Remember that the client is interested in separate recommendations for smaller airplanes and larger airplanes. Choose a passenger threshold of 20 and separate the plane types. 

```python
df['Plane.Size'] = df['Total.People'].apply(lambda x: 'Small' if x < 20 else 'Large')
```

```python
df['Plane.Size'].value_counts()
```

```python
df[['Total.People', 'Plane.Size']].head()
```

#### Analyzing Makes

Explore the human injury risk profile for small and larger Makes:
- choose the 15 makes for each group possessing the lowest mean fatal/seriously injured fraction
- plot the mean fatal/seriously injured fraction for each of these subgroups side-by-side

```python
make_stats = df.groupby('Make')['Injury.Rate'].mean().sort_values()
make_stats.head(10)
```

```python
make_counts = df['Make'].value_counts()

valid_makes = make_counts[make_counts > 20].index

make_stats = df[df['Make'].isin(valid_makes)].groupby('Make')['Injury.Rate'].mean().sort_values()

make_stats.head(10)
```

```python
make_stats.head(10).plot(kind='bar')
plt.title('Makes with lowest injury rates')
plt.ylabel('Injury Rate')
#rotating the x-axis labels for better viewing
plt.xticks(rotation=45)
plt.show()
```

```python
size_stats = df.groupby('Plane.Size')['Injury.Rate'].mean()
size_stats
```

```python
size_stats.plot(kind='bar')
plt.title('Injury Rate: Small vs Large Planes')
plt.ylabel('Injury Rate')
plt.xticks(rotation=0)
plt.show()
```

**Distribution of injury rates: small makes**

Use a violinplot to look at the distribution of the fraction of passengers serious/fatally injured for small airplane makes. Just display makes with the ten lowest mean serious/fatal injury rates.

```python

sns.boxplot(x='Plane.Size', y='Injury.Rate', data=df)
plt.title('Distribution of Injury Rates by Plane Size')
plt.show()
```

**Distribution of injury rates: large makes**

Use a stripplot to look at the distribution of the fraction of passengers serious/fatally injured for large airplane makes. Just display makes with the ten lowest mean serious/fatal injury rates.

```python
damage_stats = df.groupby('Make')['Destroyed'].mean().sort_values()
damage_stats.head(10)
```

**Evaluate the rate of aircraft destruction for both small and large aircraft by Make.** 

Sort your results and keep the lowest 15.

```python
damage_stats.head(10).plot(kind='bar')
plt.title('Makes with lowest destruction rate')
plt.ylabel('Destruction Rate')
plt.xticks(rotation=45)
plt.show()
```

### Exploring Other Variables
- Investigate how other variables effect aircraft damage and injury. You must choose **two** factors out of the following but are free to analyze more:

For each factor provide a discussion explaining your analysis with appropriate visualization / data summaries and interpreting your findings.

```python
weather_stats = df.groupby('Weather.Condition')['Injury.Rate'].mean().sort_values()
weather_stats.head()
```

```python
weather_stats.head(10).plot(kind='bar')
plt.title('Injury Rate by Weather Condition')
plt.ylabel('Injury Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
I analyzed how weather conditions affect injury rates.

Accidents that happened in poor weather conditions tend to have higher injury rates 
compared to clear weather. This suggests that weather plays an important role in 
accident severity.

Overall, weather seems to be a strong contributing factor to both accidents and 
injury outcomes.
```

```python
engine_stats = df.groupby('Engine.Type')['Injury.Rate'].mean().sort_values()
engine_stats.head()
```

```python
engine_stats.plot(kind='bar')
plt.title('Injury Rate by Engine Type')
plt.ylabel('Injury Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x='Engine.Type', y='Injury.Rate', data=df)
plt.title('Distribution of Injury Rates by Engine Type')
plt.xticks(rotation=45)
plt.show()
```
