### Cleaning and constructing Key Measurables

Injuries and robustness to destruction are a key interest point for the client. Clean and impute relevant columns and then create derived fields that best quantifies what the client wishes to track. **Use commenting or markdown to explain any cleaning assumptions as well as any derived columns you create.**

**Construct metric for fatal/serious injuries**

*Hint:* Estimate the total number of passengers on each flight. The likelihood of serious / fatal injury can be estimated as a fraction from this.

```python
df[['Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Uninjured']].head()
```

```python
df['Total.Fatal.Injuries'] = df['Total.Fatal.Injuries'].fillna(0)
df['Total.Serious.Injuries'] = df['Total.Serious.Injuries'].fillna(0)
df['Total.Uninjured'] = df['Total.Uninjured'].fillna(0)
```

```python
df['Total.People'] = df['Total.Fatal.Injuries'] + df['Total.Serious.Injuries'] + df['Total.Uninjured']
df['Injury.Rate'] = df['Serious.Fatal'] / df['Total.People']
df['Injury.Rate'] = df['Injury.Rate'].fillna(0)
```

**Aircraft.Damage**
- identify and execute any cleaning tasks
- construct a derived column tracking whether an aircraft was destroyed or not.

```python
df['Aircraft.damage'].value_counts()
```

```python
df['Destroyed'] = df['Aircraft.damage'] == 'Destroyed'
```

```python
df.head()
```

### Markdown
I created some columns to understand how serious the accidents were
and also checked if the aircraft was destroyed.

### Investigate the *Make* column
- Identify cleaning tasks here
- List cleaning tasks clearly in markdown
- Execute the cleaning tasks
- For your analysis, keep Makes with a reasonable number (you can put the threshold at 50 though lower could work as well)

```python
df['Make'].head()
```

```python
df = df[df['Make'].notna()]
```

```python
df['Make'].value_counts().head(10)
```

```python
make_counts = df['Make'].value_counts()

common_makes = make_counts[make_counts > 50].index

df = df[df['Make'].isin(common_makes)]
```

```python
df['Make'].value_counts().head()
```

### Markdown

so the analysis is more reliable.

```python
df['Model'].head()
```

```python
df['Model'].isnull().sum()
```

```python
df = df[df['Model'].notna()]
```

```python
model_counts = df['Model'].value_counts()

common_models = model_counts[model_counts > 20].index

df = df[df['Model'].isin(common_models)]
```

### Cleaning other columns
- there are other columns containing data that might be related to the outcome of an accident. We list a few here:
- Engine.Type
- Weather.Condition
- Number.of.Engines
- Purpose.of.flight
- Broad.phase.of.flight

Inspect and identify potential cleaning tasks in each of the above columns. Execute those cleaning tasks. 

**Note**: You do not necessarily need to impute or drop NaNs here.

```python
df['Engine.Type'].value_counts().head()
df['Weather.Condition'].value_counts().head()
df['Purpose.of.flight'].value_counts().head()
```

```python
df['Engine.Type'] = df['Engine.Type'].fillna('Unknown')
df['Weather.Condition'] = df['Weather.Condition'].fillna('Unknown')
df['Purpose.of.flight'] = df['Purpose.of.flight'].fillna('Unknown')
df['Broad.phase.of.flight'] = df['Broad.phase.of.flight'].fillna('Unknown')
```

```python
df['Number.of.Engines'] = df['Number.of.Engines'].fillna(df['Number.of.Engines'].median())
```

### Markdown
I filled missing values in some columns to make the data more complete.


### Column Removal

```python
df = df.drop(columns=['Latitude', 'Longitude', 'FAR.Description'])
```
